```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow # Removed extra dot from this line

from google.colab import files
uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]
image=cv2.imread(filename)
cv2_imshow(image)
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
resized=cv2.resize(image,(300,300))
cv2_imshow(resized)
```

```python
a=[image,gray]
for i in a:
  resized=cv2.resize(i,(300,300))
  cv2_imshow(resized)
```

```python
annotated=image.copy()
cv2.rectangle(annotated,(70,50),(300,300),(0,255,0),3)
cv2.putText(annotated,"My Box",(40,45),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,0,0),2)
cv2_imshow(annotated)
```

DATA AUGMENTATION


```python
cv2.imwrite('output_image.jpg',annotated)
print("Image saved as output_image.jpg")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator , load_img,img_to_array
```

```python
uploaded=files.upload()
filename=list(uploaded.keys())[0]
```

```python
img=load_img(filename,target_size=[224,224])
img_array=img_to_array(img)
img_array=np.expand_dims(img_array,axis=0)
```

```python
datagen=ImageDataGenerator(
    # rotation_range=40,
    width_shift_range=0.2,
    # height_shift_range=0.2,
    # shear_range=0.2,
    # zoom_range=0.2,
    # horizontal_flip=True,
    # brightness_range=[0.5,1.5]

)
```

```python
i=0
for batch in datagen.flow(img_array,batch_size=1):
  plt.figure(figsize=(2,2))
  plt.imshow(batch[0].astype('uint8')) # Changed 'unit8' to 'uint8'
  plt.axis('off')
  i+=1
  if i>5:
    break
```
